```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Customer Churn.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
(df.describe()).round(2)
```

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
df["TotalCharges"].dtype
```

checking if unique id is duplicated

```python
df["customerID"].duplicated().sum()
```

converted 0 and 1 values of senior citizen to yes/no to make it easier to understand

```python
def conv(value):
    if value == 1:
        return "yes"
    else : return "no"
```

```python
df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

```python
df["SeniorCitizen"]
```

```python
df["Churn"].value_counts()
```

```python
churn_counts = df["Churn"].value_counts()

plt.figure(figsize=(6,4))
plt.bar(churn_counts.index, churn_counts.values, color=['skyblue', 'salmon'])
plt.title("Customer Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby("Churn").agg({"Churn": "count"})
colors = ["#FF9A76", "#FFD93D"]
plt.pie(gb["Churn"],labels = gb.index,autopct = "%1.2f%%",colors = colors)
plt.title("percentage of Churn customer",fontsize = 10)
plt.show()
```

From the given pie chart we can see that 26.54 % of our customers has churned out

## Exploration of reasons behind churned out customers

```python
plt.figure(figsize = (5,3))
sns.countplot(x = "gender",data = df,palette=["#845EC2", "#FF9671"],hue = "Churn",legend =  False)
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize = (7,4))
ax = sns.countplot(x = "SeniorCitizen", data = df,hue="SeniorCitizen",palette=["#FF6F91", "#6A9EFF"],legend=False)
ax.bar_label(ax.containers[0])
plt.title("count of customers for seniorcitezen")
plt.show()
```

comparative a greater pecentage of people in senior citizen category have churned

```python
plt.figure(figsize = (6,3))
sns.histplot(data = df,x = "tenure",hue = "Churn",bins = 45)
plt.show()
```

from the above graph we can see that people who has used our services for a long time have stayed than the  people who has services for short term

```python
plt.figure(figsize = (7,4))
ax = sns.countplot(data = df,x = "Contract",hue = "Churn",palette = ["#00B8A9", "#F6416C"],legend = False)
ax.bar_label(ax.containers[0])
plt.title("count of customer by contract")
```

From the above graph we can see that people who have month to month contract are likely to churn then from those who have 1 or 2 years or contract. 

```python
plt.figure(figsize = (10,4))
ax = sns.countplot(x = "PaymentMethod", data = df, hue = "Churn",legend = 'auto',palette = ["#FF7E5F", "#FEB47B"])
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Churned Customers by Payment Method")
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize = (15,14))

plt.subplot(3,3,1)
plt.bar(churn_counts.index, churn_counts.values, color=['skyblue', 'salmon'])
plt.title("Customer Churn Distribution",fontsize = 10)
plt.xlabel("Churn")
plt.ylabel("Count")

plt.subplot(3,3,2)
gb = df.groupby("Churn").agg({"Churn": "count"})
colors = ["#FF9A76", "#FFD93D"]
plt.pie(gb["Churn"], labels=gb.index, autopct="%1.2f%%", colors=colors)
plt.title("percentage of Churn customer", fontsize=10)

plt.subplot(3,3,3)
sns.countplot(
    x="gender",
    data=df,
    palette=["#845EC2", "#FF9671"],
    hue="Churn",
    legend=False
)
plt.title("Churn by Gender", fontsize=10)

# -------- 4th subplot --------
plt.subplot(3,3,4)
ax = sns.countplot(
    x="SeniorCitizen",
    data=df,
    hue="SeniorCitizen",
    palette=["#FF6F91", "#6A9EFF"],
    legend=False
)
ax.bar_label(ax.containers[0])
plt.title("count of customers for seniorcitezen")

# -------- 5th subplot --------
plt.subplot(3,3,5)
ax = sns.countplot(
    x="PaymentMethod",
    data=df,
    hue="Churn",
    legend='auto',
    palette=["#FF7E5F", "#FEB47B"]
)
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Churned Customers by Payment Method")
plt.xticks(rotation=35, ha='right')
plt.tight_layout()

# -------- 6th subplot --------
plt.subplot(3,3,6)
ax = sns.countplot(
    data=df,
    x="Contract",
    hue="Churn",
    palette=["#00B8A9", "#F6416C"],
    legend=False
)
ax.bar_label(ax.containers[0])
plt.title("count of customer by contract")

plt.subplot(3,1,3)
sns.histplot(data=df, x="tenure", hue="Churn", bins=45)
plt.show()
```

# Customer Churn Analysis Summary 

#### About 26.5% of customers have churned, while 73.5% remain.
#### Gender has little effect â€” churn rates are similar for males and females.
#### Senior citizens make up a smaller portion of customers, but they churn slightly more often.
#### Customers with shorter tenure are more likely to churn.
#### Contract type strongly impacts churn â€” most churners are on month-to-month contracts, while one-year and two-year contracts show much lower churn rates.
